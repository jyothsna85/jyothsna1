### number to list conversion
- 19563212-[1,9,5,6,3,2,1,2]

```
def numbertolist(n):
  li=[]
  while n!=0:
    r=n%10
    li.append(r)
    n=n//10
  li.reverse()
  return li
numbertolist(19563212)
```

### contact  application


### module 
- add contact
- search contact
- import contact
- update the contact detail
- delete the contact

```
contacts={}
# define the function for add contacts
def addcontacts(name,phone):
  if name not in contacts:
    contacts[name]=phone
    print('contact added')
  else:
    print('contact is not added')
  return
addcontacts('xyz',1234567897)
addcontacts('aaa',7894561231)
addcontacts('xyz',1234567897)
```

```
def searchcontact(name):
  if name in contacts:
    print("name"," : ",contacts[name])
    print("contact is present")
  else:
    print('contact is not present')
  return
searchcontact('xyz')
searchcontact('aaa')
searchcontact('hhh')
```

```
# define the function for import the new contacts
def importcontacts(newcontacts):
  contacts.update(newcontacts)
  print('newcontacts')
  return
newcontacts={"ppp":1234567891,"pqr":22222222}
importcontacts(newcontacts)
```

```
# define a function for update details
def updatedetails(name,phone):
  if name in contacts:
    contacts[name]=phone
    print('contact is updated')
  else:
    print('please update the contact')
  return
updatedetails('ppp',123456895)
updatedetails('prs',789456123)
```

```
# define function to delete the contact details
def deletecontacts(name):
  if name in contacts:
    del contacts[name]
    print('contact is deleted')
  else:
    print('contact is not present')
  return
deletecontacts('aaa')
deletecontacts('bbb')
```

```
# define a function to convert the string to list
def stringtolist(s):
  li=s.split()
  numberlist=[]
  for i in li:
    numberlist.append(int(i))
  return numberlist
stringtolist('1 2 3 4 5 6')
```
